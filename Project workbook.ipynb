```python
import pandas as pd

def load_csv_file(filepath):
    df = pd.read_csv(filepath)
    return df

Disney = load_csv_file("Data - Disney.csv")
Netflix = load_csv_file("Data - Netflix.csv")
```

```python
#Give the nature of 10-K fillings, this functions get the previous 3 quarters give a quarter
def get_previous_quarters(year_quarter):
    year, quarter = year_quarter.split('-Q')
    year = int(year)
    quarter = int(quarter)

    previous_quarters = []
    for i in range(1, 4):
        if quarter == 1:
            year -= 1
            quarter = 4
        else:
            quarter -= 1

        previous_quarter = f"{year}-Q{quarter}"
        previous_quarters.append(previous_quarter)

    return previous_quarters

def mean_by_column(df, columns, columnname):
    for value in df[columnname].unique():
        mask = df[columnname] == value
        mean_values = df.loc[mask, columns].sum().div(mask.sum())
        df.loc[mask, columns] = mean_values.to_list()
    return df


def find_source_columns(df):

    source_cols = []
    for col in df.columns:
        if "(Source)" in col:
            source_cols.append(col)
    return source_cols

def get_previous_quarters_for_10k(df, program, measure):
    # Filter the DataFrame for rows where Source Type is "10-K SEC Filing"
    rows_to_be_updated = df[df['Source Type'] == '10-K SEC Filling']
    rows_to_be_updated = rows_to_be_updated[rows_to_be_updated['Program'] == program]
    # Iterate over each row in rows_to_be_updated
    for index, row in rows_to_be_updated.iterrows():
        # Get the Year-Quarter value from the row
        year_quarter = row['Year-Quarter']

        # Call the get_previous_quarters function to get the previous quarters
        previous_quarters = get_previous_quarters(year_quarter)

        # Filter the original DataFrame for the matching rows
        matching_rows = df[(df['Year-Quarter'].isin(previous_quarters)) & (df['Program'] == program)]

        # Sum the values in the "measure" column for the matching rows
        matching_sum = matching_rows[measure].sum()

        # Calculate the difference between the matching sum and the value in the "measure" column for the current row

        difference =  row[measure] - matching_sum

        # Insert the difference into the original DataFrame in the measure column
        df.at[index, measure] = difference

    return df  # Return the updated rows in the original DataFrame
```

```python
list_of_columns = ["Selling, general, administrative and other","Other operating expense"]
mean_by_column(Disney,list_of_columns,"Year-Quarter")

get_previous_quarters_for_10k(Disney, "Disney+", "Programming and production costs")
get_previous_quarters_for_10k(Disney, "Hulu SVOD", "Programming and production costs")
get_previous_quarters_for_10k(Disney, "Disney+", "Other operating expense")
get_previous_quarters_for_10k(Disney, "Hulu SVOD", "Other operating expense")

get_previous_quarters_for_10k(Disney, "Disney+", "Selling, general, administrative and other")
get_previous_quarters_for_10k(Disney, "Hulu SVOD", "Selling, general, administrative and other")

```

```python
get_previous_quarters_for_10k(Netflix, "Netflix", "Revenue")
get_previous_quarters_for_10k(Netflix, "Netflix", "Cost of Revenues")
get_previous_quarters_for_10k(Netflix, "Netflix", "Marketing")
get_previous_quarters_for_10k(Netflix, "Netflix", "Technology and development")
get_previous_quarters_for_10k(Netflix, "Netflix", "General and administrative")
```

```python
list_to_drop_from_netflix = find_source_columns(Netflix)
list_to_drop_from_disney = find_source_columns(Disney)
list_to_drop_from_both = ["Source Type","Source Url","Company"]
Netflix = Netflix.drop(list_to_drop_from_netflix,axis=1)

Disney = Disney.drop(list_to_drop_from_disney,axis=1)
```

```python
Disney["Revenue"] = Disney["Average Monthly Revenue Per Paid Subscriber"]*3*Disney["Paid memberships at end of period"]
```

```python
Netflix["Selling, general, administrative and other"] = Netflix["Marketing"] + Netflix["General and administrative"]

Netflix["Programming and production costs"] = Netflix["Cost of Revenues"]
Netflix["Other operating expense"] = Netflix["Technology and development"]
```

```python
list_to_drop_from_both = ["Source Type","Source Url","Company"]


Disney = Disney.drop(list_to_drop_from_both,axis=1)
```

```python
netflix_second_set = ["Marketing","Technology and development","General and administrative","Cost of Revenues", "Subprogram"]
Disney = Disney.drop(["Average Monthly Revenue Per Paid Subscriber"],axis=1)
Netflix =  Netflix.drop(list_to_drop_from_both,axis=1)

Netflix = Netflix.drop(netflix_second_set,axis=1)
```

```python
columns_in_thousands = ["Paid memberships at end of period","Revenue","Selling, general, administrative and other",
                       "Programming and production costs","Other operating expense"]

def divide_columns_by_1000(df, cols_to_divide):

    for col in cols_to_divide:
        df[col] = df[col] / 1000
    return df

divide_columns_by_1000(Netflix,columns_in_thousands)
```

```python
reorder = ['Program', 'Year-Quarter', 'Quarter', 'Year',"Paid memberships at end of period",
          "Revenue","Selling, general, administrative and other","Programming and production costs","Other operating expense"]

Disney = Disney.reindex(columns=reorder)
```

```python
Financial_Performance_Dataset= pd.concat([Disney, Netflix])
```

```python
Financial_Performance_Dataset.to_csv('Final_result.csv', index=False)
```
