```python
import pandas as pd

def load_csv_file(filepath):
    df = pd.read_csv(filepath)
    return df

df_1 = load_csv_file("Data.csv")
```

```python
df_1.head()
```

```python
df_1["Program_Key"] = df_1["Program"] + df_1["Quarter"]
df_1 = df_1.fillna(0)
df_1.head()
```

```python
df_1.drop(columns=['Company', 'Program',"Subprogram","Quarter"])
df_2 =  df_1.groupby('Program_Key').sum()
df_2 = df_2.reset_index()
df_1['Programming and production cost'] = df_1['Programming and production cost'].astype(float)
df_1['Revenue Per Quarter (Millions)'] = df_1['Revenue Per Quarter (Millions)'].astype(float)
df_1.dtypes
```

```python
def extract_values(string):
    application = string[:string.find('Q')].strip()
    quarter = string[string.find('Q')+1]
    year = string[-4:]
    return (application, quarter, year)

df_2[['Application', 'Quarter', 'Year']] = df_2['Program_Key'].apply(extract_values).apply(pd.Series)

df_2.drop(columns=["Program_Key"])
df_2['Year-Quarter'] = df_2['Year'] + "-"+ df_2['Quarter']
df_2.drop(columns=["Program_Key", "Year", "Quarter"])
df_2.head()
```

```python
import altair as alt
import pandas as pd


# create the stacked line chart
chart = alt.Chart(df_2).mark_line().encode(
    x=alt.X('Year-Quarter', title='Year-Quarter'),
    y=alt.Y('Paid subscribers (Millions)', title='Paid subscribers (Millions)'),
    color=alt.Color('Application', title='Application')
).properties(
    width=600,
    height=400
).configure_legend(
    titleFontSize=14,
    labelFontSize=12
)

# display the chart
chart
```
