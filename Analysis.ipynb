```python
import pandas as pd
import altair as alt

def load_csv_file(filepath):
    df = pd.read_csv(filepath)
    return df

Financial_Performance = load_csv_file("Final_result.csv")
```

```python
Financial_Performance.head()
```

```python
Financial_Performance["Cost"] =  Financial_Performance["Selling, general, administrative and other"] + Financial_Performance["Programming and production costs"] + Financial_Performance["Other operating expense"]
Financial_Performance["Revenue Label"] = Financial_Performance["Revenue"].apply(
    lambda x: f"${round(x/1000, 1)}B"  if x > 0 else ""
)


Financial_Performance["Cost Label"] = Financial_Performance["Cost"].apply(
    lambda x: f"${round(x/1000, 1)}B"  if x > 0 else ""
)
Financial_Performance["Paid memberships at end of period"] = Financial_Performance["Paid memberships at end of period"].round(0)

```

```python

base = alt.Chart(Financial_Performance).properties(width=550)

line = base.mark_line().encode(
    x='Year-Quarter',
    y='Paid memberships at end of period',
    color='Program'
)

text = alt.Chart(Financial_Performance).mark_text(
    align='right',
    baseline='middle',
    dx = 20,
    dy = -15,
    fontSize=11
).encode(
    x='Year-Quarter',
    y='Paid memberships at end of period',
    text='Paid memberships at end of period'

)

graph1 = (line + text ).properties(
     width=800,
    height=400, title={
      "text": ["Paying Subscribers  at the end of the quarter, in millions"], 
      "color": "black"

    }
)
graph1
```

```python


Financial_Performance = Financial_Performance[Financial_Performance["Year"] >= 2021]

```

```python
base = alt.Chart(Financial_Performance).properties(width=550)

line = base.mark_line(strokeWidth=2).encode(
    x='Year-Quarter',
    y='Revenue',
    color='Program'
)

line2 = base.mark_line(strokeWidth=2, strokeDash=[5, 5]).encode(
    x='Year-Quarter',
    y='Cost',
    color='Program'
)

graph2 = alt.layer(line, line2).properties(
    width=800,
    height=400,
    title={
        "text": ["Revenue Generated & Cost by Platform, in Millions of USD"],
        "color": "black"
    }
)

graph2
```
