```python
import pandas as pd

def load_csv_file(filepath):
    df = pd.read_csv(filepath)
    return df

Financial_Performance = load_csv_file("Financial Performance.csv")
```

```python
Financial_Performance.head()
```

```python
Financial_Performance["Revenue Label"] = Financial_Performance["Revenue"].apply(
    lambda x: f"${round(x/1000, 1)}B"  if x > 0 else ""
)

Financial_Performance["Paid memberships at end of period"] = Financial_Performance["Paid memberships at end of period"].round(0)

Financial_Performance["Cost"] =  Financial_Performance["Selling, general, administrative and other"] + Financial_Performance["Programming and production costs"] + Financial_Performance["Other operating expense"]
```

```python
Financial_Performance
```

```python
import altair as alt

# Define the base chart


points = alt.Chart(Financial_Performance).mark_point(filled=True, size = 100).encode(
    x='Year-Quarter',
    y='Paid memberships at end of period',
    color=alt.Color('Program'),
    size = alt.Size('Revenue')
)

text = alt.Chart(Financial_Performance).mark_text(
    align='right',
    baseline='middle',
    dx=-11,
    fontSize=11
).encode(
    x='Year-Quarter',
    y='Paid memberships at end of period',
    text='Revenue Label'

)



result = (points + text ).properties(
     width=800,
    height=400, title={
      "text": ["2018 Tourism Spend per International Visitor by New Business Density Competition for Mediterranean European Countries"], 
      "subtitle": ["New Business Density is a measure of number of new business created per year by working aged population", "Total International Visitors are Coded in Color","Total Spend by International Tourists is coded in size"],
      "color": "black",
      "subtitleColor": "grey"
    }
)
result
```

```python

base = alt.Chart(Financial_Performance).properties(width=550)

line = base.mark_line().encode(
    x='Year-Quarter',
    y='Paid memberships at end of period',
    color='Program'
)

text = alt.Chart(Financial_Performance).mark_text(
    align='right',
    baseline='middle',
    dx = 20,
    dy = -15,
    fontSize=11
).encode(
    x='Year-Quarter',
    y='Paid memberships at end of period',
    text='Paid memberships at end of period'

)

result = (line + text ).properties(
     width=800,
    height=400, title={
      "text": ["Paying Subscribers  at the end of the quarter, in millions"], 
      "color": "black"

    }
)
result
```
